```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/home/brainstimmap/Documents/DATA/20xx_Projects/2025_DBSinDepression/03_Data/Patient3/VTA_tracts/plot/structural/data4heatmap.csv'
assessment_list = ['BDI', 'HAMD', 'MADRS', 'SHAPS', 'Sheehan', 'SOFAS', 'BARS']
complete_df = pd.read_csv(path)

```

```python
assessment_list +['date']
```

```python
for measure in assessment_list:
    selection = complete_df[[c for c in complete_df.columns if c not in assessment_list+ ['date']]+[measure]]
    
    to_plot_df = selection.sort_values(by=measure,ascending=False).set_index(measure)
    fig, axes = plt.subplots(1, 2,figsize=(20,8))
    plt.suptitle(measure)
    sns.heatmap( to_plot_df.transpose(),linewidths=.5,ax=axes[0])    
    sns.heatmap(to_plot_df.drop(['BA25'],axis=1).transpose(),linewidths=.5,ax=axes[1])
    plt.show()
    
```
